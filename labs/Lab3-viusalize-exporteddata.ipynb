```python
###### Import necessary libraries
from IPython.core.debugger import set_trace
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)

plt.style.use(style="seaborn")
%matplotlib inline
```

#### Import data

```python
df = pd.read_csv(f"../dataset/021_1min_seg.csv")
df.shape
```

```python
####### Visualize

# draw line plot of amount
f = plt.figure(1, figsize=(16, 20))

i=1
for string in df.keys():
    if (string not in ['local_time']):
        axis = f.add_subplot(len(df.keys()),2,i)
        axis.plot(df[string])
        axis.title.set_text(string + " - line plot")
    
        axis = f.add_subplot(len(df.keys()),2,i+1)
        axis.boxplot(df[string])
        axis.title.set_text(string + " - Box plot")
        i+=2

plt.show()
```
