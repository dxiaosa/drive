#### Visulization
line plot -> global and local treand

boxplot -> data distribution. 

To know about the boxplolt, please refer this link: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51

```python
####### Visualize

# draw line plot of amount
f = plt.figure(1, figsize=(16, 20))

i=1
for string in df.keys():
    if (string not in ['local_time']):
        # we plot the line chart to help observe the trend of each time sereis
        axis = f.add_subplot(len(df.keys()),2,i)
        axis.plot(df[string])
        axis.title.set_text(string + " - line plot")
        
        # we also show the boxplot to impress the data distribution         
        axis = f.add_subplot(len(df.keys()),2,i+1)
        sns.boxplot(data=df[string], ax=axis)
        axis.title.set_text(string + " - Box plot")
        i+=2

plt.show()
```
