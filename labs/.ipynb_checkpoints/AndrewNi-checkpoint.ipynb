```python
###### Import necessary libraries
from IPython.core.debugger import set_trace
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)

plt.style.use(style="seaborn")
%matplotlib inline
```

#### Data preparation

Synthetic Financial Datasets For Fraud Detection

It simulates mobile money transactions based on a sample of real transactions extracted from one month of financial logs from a mobile money service implemented in an African country. 

Src: https://www.kaggle.com/datasets/ealaxi/paysim1

```python
df = pd.read_csv(
    f"PS_20174392719_1491204439457_log.csv", 
    encoding="ISO-8859-1", 
    error_bad_lines=False
)
df.shape
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
####### Resample 'amount' field with interval=10000
amount = df['amount'].to_frame()
subsample = amount[::10000].reset_index(drop=True)
subsample

#Data features to visualize and analyze
visualized_data = {'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest'}
array = [string for string in df.keys() if not string in visualized_data]

#Dropping unnecessary columns and subsampling data
df_subsample = df.copy().drop(columns=array)[::10000].reset_index(drop=True)
```

```python
####### Visualize


# draw line plot of amount
f = plt.figure(1, figsize=(16, 20))

i=1
for string in df_subsample.keys():
    if (string != "iforest"):
        axis = f.add_subplot(len(df_subsample.keys()),2,i)
        axis.plot(df_subsample[string])
        axis.title.set_text(string + " - line plot")
    
        axis = f.add_subplot(len(df_subsample.keys()),2,i+1)
        _ = axis.boxplot(df_subsample[string])
        axis.title.set_text(string + " - Box plot")
        i+=2

plt.show()
```

```python
####### iForest train

from sklearn.ensemble import IsolationForest

# model one for single variable data
model_one = IsolationForest(
    contamination=0.01, 
    n_estimators=100
)
# model two for multivariate data
model_two = IsolationForest(
    contamination=0.01, 
    n_estimators=100
)

# train
model_one.fit(subsample)
model_two.fit(df_subsample)
```

```python
####### Prediction

#Not sure why df_subsample wouldn't work with pd.Series() like for subsample
df_subsample["iforest"] = model_two.fit(df_subsample).predict(df_subsample)
df_subsample["iforest"] = df_subsample["iforest"].map({1:0, -1:1})
print(df_subsample)

subsample["iforest"] = pd.Series(model_one.predict(subsample))
subsample["iforest"] = subsample["iforest"].map({1: 0, -1: 1})
```

```python
####### Visualize anomalies
from matplotlib.collections import EllipseCollection


### Line plot and anomalies for multivariate iForest
f = plt.figure(1, figsize=(16, 24))
anomalies = df_subsample[df_subsample["iforest"]==1]
i=1

for string in df_subsample.keys():
    if (string != "iforest"):
        axis = f.add_subplot(len(df_subsample.keys())+1,1,i)
        axis.plot(df_subsample[string])
        axis.title.set_text(string)
        
        #anomalies
        axis.scatter(
            anomalies.index,
            anomalies[string],
            marker='o',
            c="red",
            s=40
        )
        i+=1


####### draw line plot of amount

ax = f.add_subplot(len(df_subsample.keys())+1,1,len(df_subsample.keys())+1)
ax.plot(subsample)
ax.title.set_text("Amount")

####### plot anomaly with marker


anomaly = subsample[subsample['iforest']==1]

ax.scatter(
    anomaly.index,
    anomaly['amount'],
    marker='o',
    c='red',
    s=40
)

plt.show()

print("""Multivariate anomaly detection is presented in the top five graphs and single-variable detection on the bottom graph. It seems like the number of anomalies is roughly the same between the multivariate and single variable anomaly detection methods, probably because they have identical contamination parameters. However, they identify different transactions as anomalous.""")
```

```python
##### Plot the boxplot after removing the extreme anomaleis

df_normal = df_subsample[df_subsample["iforest"]==0]
i=1
f = plt.figure(1, figsize=(16, 20))

for string in df_normal.keys():
    if (string != "iforest"):
        axis = f.add_subplot(len(df_normal.keys())+1,1,i)
        _ = axis.boxplot(df_subsample[string])
        axis.title.set_text(string + " - Box plot")
        i+=1


normal = subsample[subsample['iforest']==0]


ax = f.add_subplot(len(df_normal.keys())+1,1,len(df_normal.keys())+1)
ax.boxplot(normal['amount'])
plt.show()
```
